```python
def update(rows):
    trv.delete(*trv.get_children())
    new_l = []
    for i in range(len(rows)):
        for j in sd:
            new_l.append(sd[j][i])
        trv.insert('',"end",values=new_l,tag='gray')
        new_l = []

'''def update(data):
    trv.insert('',"end",values=['0','1','2','3'])
'''

def search_data():
    search_entry = ent.get()
    trv.delete(*trv.get_children())
    new_l = []
    
    file_path = label_file["text"]
    try:
        excel_filename = r"{}".format(file_path)
        if excel_filename[-4:] == ".csv":
            df = pd.read_csv(excel_filename)
        else:
            df = pd.read_excel(excel_filename)

    except ValueError:
        tk.messagebox.showerror("Information", "The file you have chosen is invalid")
        return None
    except FileNotFoundError:
        tk.messagebox.showerror("Information", f"No such file as {file_path}")
        return None

    clear_data()
    trv["column"] = list(df.columns)
    trv["show"] = "headings"
    for column in trv["columns"]:
        trv.heading(column, text=column,anchor=CENTER) # let the column heading = column name
        trv.column(column,anchor=CENTER)
       
    for i in range(len(df)):
        for j in sd:
            new_l.append(df[j][i])
        if search_entry in new_l:
            trv.insert('',"end",values=new_l,tag='gray')
        new_l = []   
        
        
def clear_button():
    ent.delete(0,END)
    Load_excel_data()
    
def get_row(event):
    rowid = trv.identify_row(event.y)
    item = trv.item(trv.focus())
    en1.set(item['values'][0])
    en2.set(item['values'][1])
    en3.set(item['values'][2])
    en4.set(item['values'][3])
    pass

def update_entry():
    messagebox.showerror(title="Permission Denied !", message="You do not have enough permission to perform this task")
#     if messagebox.askyesno("Caution !","Are you sure want to update this entry?"):
#         #sd = pd.read_csv("data/student_data.csv")
#         file_path = label_file["text"]
#         try:
#             excel_filename = r"{}".format(file_path)
#             if excel_filename[-4:] == ".csv":
#                 df = pd.read_csv(excel_filename)
#             else:
#                 df = pd.read_excel(excel_filename)

#         except ValueError:
#             tk.messagebox.showerror("Information", "The file you have chosen is invalid")
#             return None
#         except FileNotFoundError:
#             tk.messagebox.showerror("Information", f"No such file as {file_path}")
#             return None
        
#         old_student_id = en1.get()
#         old_fname = en2.get()
#         old_lname = en3.get()
#         old_stream = en4.get()
#         new_studentid = askstring('Student ID', 'What is the Student ID?')
#         new_fname = askstring('First Name', 'What is his First Name?')
#         new_lname = askstring('Last Name', 'What is his Last Name?')
#         new_stream = askstring('Stream', 'What is his Stream?')
        
#         df = df.replace(old_student_id,new_studentid)
#         df = df.replace(old_fname,new_fname)
#         df = df.replace(old_lname,new_lname)
#         df = df.replace(old_stream,new_stream)
#         df.to_csv(file_path, index=False)
        
#     else:
#         pass

def add_entry():
    messagebox.showerror(title="Permission Denied !", message="You do not have enough permission to perform this task")
#     #print(sd)
#     if messagebox.askyesno("Confirm adding the entry?","Are you sure want to add this entry?"):
#         file_path = label_file["text"]
#         try:
#             excel_filename = r"{}".format(file_path)
#             if excel_filename[-4:] == ".csv":
#                 df = pd.read_csv(excel_filename)
#             else:
#                 df = pd.read_excel(excel_filename)

#         except ValueError:
#             tk.messagebox.showerror("Information", "The file you have chosen is invalid")
#             return None
#         except FileNotFoundError:
#             tk.messagebox.showerror("Information", f"No such file as {file_path}")
#             return None
#         student_id = en1.get()
#         fname = en2.get()
#         lname = en3.get()
#         stream = en4.get()
#         data = {
#             'Student ID':[student_id],
#             'First Name':[fname],
#             'Last Name':[lname],
#             'Stream':[stream],
#         }
#         df2 = pd.DataFrame(data)
#         df = df.append(df2,ignore_index = True)
#         df.to_csv(file_path, index=False)
#     else:
#         pass

def delete_entry():
    messagebox.showerror(title="Permission Denied !", message="You do not have enough permission to perform this task")
#     if messagebox.askyesno("Confirm Delete?","Are you sure want to delete this entry?"):

#         file_path = label_file["text"]
#         try:
#             excel_filename = r"{}".format(file_path)
#             if excel_filename[-4:] == ".csv":
#                 df = pd.read_csv(excel_filename)
#             else:
#                 df = pd.read_excel(excel_filename)

#         except ValueError:
#             tk.messagebox.showerror("Information", "The file you have chosen is invalid")
#             return None
#         except FileNotFoundError:
#             tk.messagebox.showerror("Information", f"No such file as {file_path}")
#             return None
        
#         student_id = en1.get()
#         sd2 = pd.read_csv(excel_filename,index_col=False)
#         sd2 = sd2[sd2['Student ID']!=student_id]
#         sd2.to_csv(file_path, index=False)
#     else:
#         pass
    
def File_dialog():
    """This Function will open the file explorer and assign the chosen file path to label_file"""
    filename = filedialog.askopenfilename(initialdir="/",
                                          title="Select A File",
                                          filetype=(("xlsx files", "*.xlsx"),("All Files", "*.*")))
    label_file["text"] = filename
    return None

def Load_excel_data():
    """If the file selected is valid this will load the file into the Treeview"""
    file_path = label_file["text"]
    try:
        excel_filename = r"{}".format(file_path)
        if excel_filename[-4:] == ".csv":
            df = pd.read_csv(excel_filename)
        else:
            df = pd.read_excel(excel_filename)

    except ValueError:
        tk.messagebox.showerror("Information", "The file you have chosen is invalid")
        return None
    except FileNotFoundError:
        tk.messagebox.showerror("Information", f"No such file as {file_path}")
        return None
    
    #Creation of Striped rows
    trv.tag_configure("oddrow",background="white")
    trv.tag_configure("evenrow",background="#d9d9d9")
    
    #Uploading data
    clear_data()
    trv["column"] = list(df.columns)
    trv["show"] = "headings"
    for column in trv["columns"]:
        trv.heading(column, text=column,anchor=CENTER) # let the column heading = column name
        trv.column(column,anchor=CENTER)
        
    df_rows = df.to_numpy().tolist() # turns the dataframe into a list of lists
    
    global COUNT
    for row in df_rows:
        if COUNT%2==0:
            trv.insert("", "end", values=row, tags=('evenrow',)) # inserts each list into the treeview. For parameters see https://docs.python.org/3/library/tkinter.ttk.html#tkinter.ttk.Treeview.insert
        else:
            trv.insert("", "end", values=row, tags=('oddrow',))
        COUNT+=1
    
    return None

def clear_data():
    trv.delete(*trv.get_children())
    return None
```
